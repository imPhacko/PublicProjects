```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("/content/drive/MyDrive/DATA AI/car.data", header=None)

print(data)
car1 = data[data[6] == 'unacc']
car2 = data[data[6] == 'acc']
car3 = data[data[6] == 'good']
car4 = data[data[6] == 'vgood']
input_attributes = [0, 1, 2, 3, 4, 5]

#plt.scatter(car1[0], car1[3], label="Car1 type unacceptable")
#plt.scatter(car2[0], car2[3], label="Car2 type acceptable")
#plt.scatter(car3[0], car3[3], label="Car3 type good")
#plt.scatter(car4[0], car4[3], label="Car4 type very good")
plt.hist(car1[0], label="Car1 - unacceptable", histtype='bar')
plt.hist(car2[0], label="Car2 - acceptable", histtype='bar')
plt.hist(car3[0], label="Car2 - good", histtype='bar')
plt.hist(car4[0], label="Car2 - very good", histtype='bar')

plt.title("Four types of cars")

plt.xlabel("Buying cost")
plt.ylabel("Counts")
plt.legend()

```
